## Nhận xét và giải thích kết quả chạy

- Dùng SMEM chạy nhanh hơn so với khi dùng GMEM: \\
    Vì toàn bộ dữ liệu của một block, kể cả phần bù, được sao chép vào SMEM trước khi thực hiện tích chập, sau đó các thread trong cùng block sẽ dùng chung dữ liệu này. Mà tốc độ đọc/ghi với SMEM nhanh hơn GMEM. Bên cạnh đó, `SMEM / one SM` tối đa là `65536 bytes`, mà mỗi block chỉ dùng tối đa `4800 bytes SMEM` (`(32 + 8) x (32 + 8) x 3`, khi `blockSize = 32 x 32`, mỗi phần tử `uchar3` có kích thước `24 bits = 3 bytes`) nên không ảnh hưởng đến `Occupancy`.

- Dùng CMEM + SMEM chạy nhanh hơn so với khi chỉ dùng SMEM: \\
    Khi khai báo bộ lọc ở tầm vực toàn cục và lưu nó vào CMEM thì các thread chỉ cần đọc dữ liệu này từ `Const Cache` của SM. Mà tốc độ đọc từ `Const Cache` nhanh hơn so với `DRAM`.

## Giải thích thêm về mã nguồn

Tuy mã nguồn đã viết không `hard-code` giá trị nào nhưng nó được tối ưu cho ảnh vuông. Do đó, nếu dùng mã nguồn này với ảnh hình chữ nhật thì sẽ chỉ đúng với một vài `blockSize` (xem các cell code bên dưới).

Thuật toán ghi dữ liệu lên SMEM như sau: mỗi thread ngoài việc sao chép dữ liệu của chính nó vào SMEM thì còn chịu trách nhiệm kiểm tra và sao chép dữ liệu cho 4 điểm khác theo đường chéo trong bán kính `filterWidth / 2` (trên bên trái, trên bên phải, dưới bên trái, dưới bên phải). Mỗi thread cần trải qua 4 lần kiểm tra điều kiện thì mới hoàn thành quá trình sao chép dữ liệu.
