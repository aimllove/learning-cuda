## Câu 1

### 1.1 Biên dịch mã nguồn

```
!nvcc bt01_p1.cu -o bt01_p1
```

Cell biên dịch mã nguồn không báo lỗi. Như vậy chương trình đã được biên dịch thành công và được lưu thành file thực thi `bt01_p1`.

### 1.2 Thực thi chương trình với các `blockSize` khác nhau

#### a) `blockSize` 32 x 32 (mặc định)

```
!./bt01_p1 in.pnm out.pnm
```

Chương trình thực thi thành công. Ta thấy `device` chạy nhanh hơn `host` khoảng 2 lần. Độ lỗi cũng rất nhỏ, rơi vào 0.002674. So sánh với hướng dẫn của thầy thì con số này khá nhỏ (*0.xxx*) nên ta chấp nhận được.

#### b) `blockSize` 16 x 16

```
!./bt01_p1 in.pnm out.pnm 16 16
```

Chương trình thực thi thành công. Ta thấy `device` chạy nhanh hơn `host` khoảng 2 lần. Độ lỗi tương tự với `blockSize` 32x32 nên cũng chấp nhận được.

#### c) blockSize 64 x 64

```
!./bt01_p1 in.pnm out.pnm 64 64
```

<a id='explain'></a>
Chương trình thực thi thành công, nhưng **báo lỗi khi chạy kernel**. Lỗi được trả về là `invalid configuration argument` - tham số cấu hình không hợp lệ. Việc chỉ định `blockSize` 64 x 64 nghĩa là ta yêu cầu `device` cung cấp các `block` có 64 x 64 = 4096 `thread`. Mà trong [bảng này](https://docs.nvidia.com/cuda/cuda-c-programming-guide/index.html#features-and-technical-specifications__technical-specifications-per-compute-capability), dòng **Maximum number of threads per block** (cho biết số thread tối đa trên một block) của tất cả *Compute Capability* hiện tại chỉ là 1024.

Do đó kernel không thể thực thi.

Độ lỗi có một chút khác biệt:

- khi chạy bằng `device`, độ lỗi nhích lên một chút, rơi vào 0.000703. So sánh với hướng dẫn của thầy thì con số này nhỏ hơn nhiều so với 0.xxx, nên đây là một kết quả chấp nhận được.

#### b) `blockSize` 16 x 16

```
!./bt01_p2 in.pnm out2.pnm out2_target.pnm 16 16
```

Chương trình thực thi thành công. Ta thấy `device` chạy nhanh hơn `host` gần 300 lần. Độ lỗi tương tự với `blockSize` 32x32 nên chấp nhận được.

#### c) blockSize 64 x 64

```
!./bt01_p2 in.pnm out2.pnm out2_target.pnm 64 64
```

Chương trình thực thi thành công, nhưng **báo lỗi khi chạy kernel**. Lỗi được trả về là `invalid configuration argument` - tham số cấu hình không hợp lệ. Ta lý giải tương tự mục [1.2c](#explain)
